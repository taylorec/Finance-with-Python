# Future Value

$FV = PV(1+R)^n$

* FV - future value
 * future value is the value of a current asset at a specified date in the future based on an assumed rate of growth over time
* PV - present value
* R - interest rate 
* n - number of years

# Present Value

$PV = \frac{FV}{(1+R)^n}$

* FV - future value
* PV - present value
* R - interest rate 
* n - number of years

```python
# Present value of receiving $100 one year from now with an interest rate of 10%
abs(round(npf.pv(0.1,1,100),2))

#input format for fv: npf.fv(rate,nper,pv)
```

# Continuous Rate of Return

* FV = PV*$e^{r*t}$
* PV = FV$e^{-r*t}$

```python
from math import exp

def future_continuous_value(x, r, t):
    return x*exp(r*t)


def present_continuous_value(x, r, t):
    return x*exp(-r*t)
```

```python
# value of investment in dollars
x = 100
# define the interest rate (r)
r = 0.05
# duration (years)
n = 5
```

```python
print("Future value (continuous model) of x: %s" % round(future_continuous_value(x, r, n),2))
print("Present values (continuous model) of x: %s" % round(present_continuous_value(x, r, n),2))
```

# Future Value of Annuity

 FV(annuity due) = $\frac{c}{R}[(1+R)^n-1]*(1+R)$

* R - period rate
* n - number of periods
* c - constant cash flow

# Present Value of Annuity

PV(annuity due) = $\frac{c}{R}[1-\frac{1}{(1+R)^n}](1+R)$

* R - period rate
* n - number of periods
* c - constant cash flow

```python
# PV with discount rate of 1% per year and annual cash flow of $20 for the next 10 years
abs(round(npf.pv(0.01,10,20,0,1),2))

#input format for pv of annuity: npf.fv(rate,nper,cf,0,1)
```
